```python
sorted_dining= df.sort_values('Dining Rating', ascending=False)
sorted_dining
```

```python
top_restaurants = df.sort_values('Dining Rating', ascending=False).head(10)

plt.figure(figsize=(8,5))
sns.barplot(x='Dining Rating', y='Restaurant Name', data=top_restaurants)
plt.title("Top 10 Restaurants by Dining Rating")
plt.show()
```

```python
plt.figure(figsize=(8,5))
city_rating = df.groupby('City')['Dining Rating'].mean().sort_values(ascending=False).head(10)

sns.barplot(x=city_rating.values, y=city_rating.index)
plt.title("Top Cities by Average Dining Rating")
plt.xlabel("Average Rating")
plt.ylabel("City")
plt.show()
```
